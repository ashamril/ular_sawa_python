**iteration all days between 2 dates**

```
from datetime import date, timedelta

months_name = {
  "01": "januari",
  "02": "februari",
  "03": "mac",
  "04": "april",
  "05": "mei",
  "06": "jun",
  "07": "julai",
  "08": "ogos",
  "09": "september",
  "10": "oktober",
  "11": "november",
  "12": "disember"}

start_date = date(2021, 2, 20)
end_date = date(2021, 3, 10)
delta = timedelta(days=1)
while start_date <= end_date:
    #print (start_date.strftime("%Y-%m-%d"))
    month_number = start_date.strftime("%m")
    month_name = months_name[month_number]
    print (start_date.strftime("%Y {}/{} %d".format(month_name, month_number)))
    start_date += delta
```

**iteration all days between 2 years**

```
def leap_year(this_year):
    if (this_year % 4) == 0:
        if (this_year % 100) == 0:
            if (this_year % 400) == 0:
                return True

    return False


for my_year in range(2020, 2020 + 1):
    print(my_year)
    year_months = {
        1: 31,
        2: 29 if leap_year(my_year) else 28,
        3: 31,
        4: 30,
        5: 31,
        6: 30,
        7: 31,
        8: 31,
        9: 30,
        10: 31,
        11: 30,
        12: 31,
    }
    for my_month in year_months.keys():
        day = year_months[my_month]
        for my_day in range(1, day + 1):
            print("{0} {1} {2}".format(my_year, my_month, my_day))
```
