```python
import PyPDF2
```

```python
pdfFileObj = open(file_path, 'rb')
pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
```

```python
number_of_pages = pdfReader.numPages
number_of_pages
```

```python
textPDF = ''
for i in range(0 , number_of_pages):  
        pageObj = pdfReader.getPage(i)
        textPDF += pageObj.extractText()
```

```python
textPDF
```

```python
import matplotlib.pyplot as plt
from collections import Counter
from wordcloud import WordCloud, STOPWORDS
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
stopwords = nltk.corpus.stopwords.words('english')
#newStopWords = ['gmail', 'march', 'com', 'ashamril']
#stopwords.extend(newStopWords)

#word_could_dict=Counter(textPDF)
wordcloud = WordCloud(stopwords = stopwords, width=1600, height=800, background_color ='white').generate(textPDF)

plt.figure(figsize=(20, 20), facecolor='k')
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.tight_layout(pad=0)
#plt.savefig('wordcloud.png', facecolor='k', bbox_inches='tight')
plt.show()
```
