## 3.1 Auto ARIMA
ARIMA (Autoregressive integrated moving average) is a very popular statistical method for Time Series forecasting. ARIMA models take into account the past values to predict the future values. 

There are three important parameters in ARIMA:

#### p (past values used for forecasting the next value)
#### q (past forecast errors used to predict the future values)
#### d (order of differencing)

```python
df_ALL_list.head(5)
```

```python
# Select specific company into new df
# df_STOCK = df_ALL_list.loc[:, ['DATE', 'AAPL']]
# df_STOCK = df_ALL_list.loc[:, ['DATE', 'AMZN']]
# df_STOCK = df_ALL_list.loc[:, ['DATE', 'FB']]
df_STOCK = df_ALL_list.loc[:, ['DATE', 'NDAQ']]
# df_STOCK = df_ALL_list.loc[:, ['DATE', 'NKE']]
df_STOCK.head(10)
```

```python
# Indexing with Time Series Data
df_STOCK = df_STOCK.set_index('DATE')
df_STOCK.index
```

```python
data = df_STOCK.sort_index(ascending=True, axis=0)

# Divide into train and validation set (70:30)
train = data[:int(0.7*(len(data)))]
valid = data[int(0.7*(len(data))):]

# Plotting the data
train.plot()
valid.plot()
plt.title('Stock Price')
plt.xlabel('Time')
plt.ylabel('Stock Price')
plt.legend()
plt.show()
```

```python
# Building the model
# Parameter tuning for ARIMA consumes a lot of time, in this case auto ARIMA was used to automatically choose 
# the best combination of (p,q,d) that provides the least error. 
from pmdarima.arima import auto_arima
model = auto_arima(train, trace=True, error_action='ignore', suppress_warnings=True)
model.fit(train)

forecast = model.predict(n_periods=len(valid))
forecast = pd.DataFrame(forecast,index = valid.index,columns=['Prediction'])
```

Auto ARIMA takes into account the AIC and BIC values generated (as in the code) to determine the best combination of parameters. AIC (Akaike Information Criterion) and BIC (Bayesian Information Criterion) values are estimators to compare models. The lower these values, the better is the model.

```python
model.fit(train)
```

```python
forecast.head(10)
```

```python
# Plot the predictions for validation set
plt.plot(train, label='Train')
plt.plot(valid, label='Valid')
plt.plot(forecast, label='Prediction')
plt.title('Stock Price Prediction')
plt.xlabel('Time')
plt.ylabel('Stock Price')
plt.legend()
plt.show()
```

```python
# Calculate rmse
from math import sqrt
from sklearn.metrics import mean_squared_error

rmse = sqrt(mean_squared_error(valid,forecast))
print(rmse)
```

The root-mean-squared error (RMSE) is a measure of how well the model performed. It does this by measuring difference between predicted values and the actual values. 
Eg. If model predicts 10, but the actual value is 5, this difference between the prediction (10) and the actual observation (5) is the error term: (y_prediction - y_actual).
The error term is important because we usually want to minimize the error. In other words, the predictions are very close to the actual values.

## 3.2 Long Short Term Memory (LSTM)
Long Short Term Memory (LSTM) is an artificial recurrent neural network (RNN) architecture used in the field of deep learning. LSTMs are widely used for sequence prediction problems and have proven to be extremely effective. The reason they work so well is because LSTM is able to store past information that is important, and forget the information that is not. 

```python
# Read the stock market price from csv
#df = pd.read_csv("./CSV/AAPL-v3.csv", sep=',')
#df = pd.read_csv("./CSV/AMZN-v3.csv", sep=',')
#df = pd.read_csv("./CSV/FB-v3.csv", sep=',')
df = pd.read_csv("./CSV/NDAQ-v3.csv", sep=',')
#df = pd.read_csv("./CSV/NKE-v3.csv", sep=',')

# Importing required libraries
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Creating new dataframe named data and new_data
data = df.sort_index(ascending=True, axis=0)
new_data = pd.DataFrame(index=range(0,len(data)),columns=['Date', 'Close'])
for i in range(0,len(data)):
    new_data['Date'][i] = data['Date'][i]
    new_data['Close'][i] = data['Close'][i]

# Setting index
new_data.index = new_data.Date
new_data.drop('Date', axis=1, inplace=True)

# Creating train and test sets (70:30)
dataset = new_data.values
train = dataset[:int(0.7*(len(dataset)))]
valid = dataset[int(0.7*(len(dataset))):]

# We have to scale the data for optimal performance. In this case, well use Scikit- Learns MinMaxScaler 
# and scale the dataset to numbers between 0 and 1.
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(dataset)

# Creating Data with Timesteps
# LSTMs expect the data to be in a specific format, usually a 3D array. 
# To create the data in 60 timesteps and converting it into an array using NumPy. 
# Thus, converting the data into a 3D dimension array with X_train samples, 60 timestamps (To make predictions 
# only for the next day, the validation set should have only 1 row (with past 60 values), 
# and one feature at each step.
x_train, y_train = [], []
for i in range(60,len(train)):
    x_train.append(scaled_data[i-60:i,0])
    y_train.append(scaled_data[i,0])
x_train, y_train = np.array(x_train), np.array(y_train)
x_train = np.reshape(x_train, (x_train.shape[0],x_train.shape[1],1))

# Building the LSTM
# In order to build the LSTM, we need to import a couple of modules from Keras:
#    Sequential for initializing the neural network
#    Dense for adding a densely connected neural network layer
#    LSTM for adding the Long Short-Term Memory layer
#    Dropout for adding dropout layers that prevent overfitting 
from keras.models import Sequential
from keras.layers import Dense, Dropout, LSTM

# Add the LSTM layer. We add the LSTM layer with the following arguments:
#    50 units which is the dimensionality of the output space
#    return_sequences=True which determines whether to return the last output in the output sequence, 
# or the full sequence
#    input_shape as the shape of our training set. 
# Add the Dense layer that specifies the output of 1 unit then compile the model using the popular adam optimizer 
# and set the loss as the mean_squarred_error
# Next, fit the model to run on 1 epochs with a batch size of 1.
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(x_train.shape[1],1)))
model.add(LSTM(units=50))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')
model.fit(x_train, y_train, epochs=1, batch_size=1, verbose=2)

# Predicting the values, using past 60 from the train data
inputs = new_data[len(new_data) - len(valid) - 60:].values
inputs = inputs.reshape(-1,1)
inputs = scaler.transform(inputs)

# In order to predict future stock prices:
#    Merge the training set and the test set on the 0 axis.
#    Set the time step as 60 
#    Reshape the dataset as done previously 
# After making the predictions we use inverse_transform to get back the stock prices in normal readable format.
X_test = []
for i in range(60,inputs.shape[0]):
    X_test.append(inputs[i-60:i,0])
X_test = np.array(X_test)
X_test = np.reshape(X_test, (X_test.shape[0],X_test.shape[1],1))
closing_price = model.predict(X_test)
closing_price = scaler.inverse_transform(closing_price)
# Time taken = 15 minutes
```

```python
# Predicted closing price
closing_price
```

```python
# Calculate rmse
rms = np.sqrt(np.mean(np.power((valid-closing_price),2)))
rms
```

```python
# Plotting the Results
# Use Matplotlib to visualize the result of the predicted stock price and the real stock price.
train = new_data[:int(0.7*(len(new_data)))]
valid = new_data[int(0.7*(len(new_data))):]
valid['Predictions'] = closing_price
plt.plot(train['Close'], label = 'Train')
# plt.plot(valid[['Close','Predictions']])
plt.plot(valid['Close'], label = 'Valid')
plt.plot(valid['Predictions'], label = 'Prediction')
plt.title('Stock Price Prediction')
plt.xlabel('Time')
plt.ylabel('Stock Price')
plt.legend()
plt.show()
# Time taken = 5 minutes
```

From the plot above, it's clearly shows that LSTM model is predicting better than Auto ARIMA. 
And this model still can be tuned up on various parameters such as changing the number of LSTM layers, adding dropout value or increasing the number of epochs to get better and more accurate prediction.
