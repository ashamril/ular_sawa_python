## Scrapping Shopee data and compare the pricing

```
import requests
import pandas as pd

Shopee_url = "https://shopee.com.my"
keyword_search = "mountain bike"

headers = {
    'User-Agent': 'Chrome', 
    'Referer': '{}search?keyword={}'.format(Shopee_url, keyword_search)}

url = 'https://shopee.com.my/api/v2/search_items/?by=relevancy&keyword={}&limit=100&newest=0&order=desc&page_type=search'.format(keyword_search)
#url = 'https://shopee.com.my/search?keyword={}&page=0&sortBy=relevancy'.format(keyword_search)

# Shopee API request
r = requests.get(url, headers = headers).json()
```

```
pd.options.display.max_columns = None
pd.set_option('display.expand_frame_repr', False)
pd.set_option('display.max_colwidth', 1000)
```

```
# Shopee scraping script
shopid = []
itemid = []
name = []
shop_location = []
price_min_before_discount = []
price_min = []
show_discount = []
historical_sold = []
is_official_shop = []
shopee_verified = []
stock = []


for item in r['items']:
    shopid.append(item['shopid'])
    itemid.append(item['itemid'])
    name.append(item['name'])
    shop_location.append(item['shop_location'])
    price_min_before_discount.append(item['price_min_before_discount'])
    price_min.append(item['price_min'])
    show_discount.append(item['show_discount'])
    historical_sold.append(item['historical_sold'])
    is_official_shop.append(item['is_official_shop'])
    shopee_verified.append(item['shopee_verified'])
    stock.append(item['stock'])
```

```
Shopee = pd.DataFrame(zip(shopid, itemid, name, shop_location, price_min_before_discount, price_min, show_discount, historical_sold, is_official_shop, shopee_verified, stock), 
                      columns=['shopid', 'itemid', 'name', 'location', 'before_discount', 'price', 'discount_perc', 'sold', 'official_shop', 'verified', 'stock'])
df = Shopee.copy()
df
```

```
import numpy as np

#spec_chars = ["!",'"',"#","%","&","'","(",")","*","+",",","-",".","/",":",";","<","=",">","?","@","[","\\","]","^","_","`","{","|","}","~","‚Äì"]
#spec_chars = ["!",'"',"#","%","&","'","*","+",",","-",".","/",":",";","<","=",">","?","@","[","\\","]","^","_","`","{","|","}","~","‚Äì"]
spec_chars = [",","%"]

df['before_discount'] = round(df['before_discount'] / 100000, 2)
df['price'] = round(df['price'] / 100000, 2)
df['link'] = df['name']
df['link'] = df['link'].replace(" ", "-", regex=True)
for char in spec_chars:
    df['link'] = df['link'].str.replace(char, "", regex=True)
```

```
df['link'] = df['link'] + "-i." 
df['link'] = df['link'] + df['shopid'].map(str) + "." 
df['link'] = "https://shopee.com.my/" + df['link'].map(str) + df['itemid'].map(str)
df = df.drop(['itemid'], axis=1)
df
```
